```python
# read in review file
import pandas as pd
import gzip
import json


def parse(path):
    g = gzip.open(path, "rb")
    for l in g:
        yield json.loads(l)


def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient="index")
```

```python
from collections import defaultdict

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from adjustText import adjust_text
from scipy.sparse import csr_matrix
from cornac.eval_methods import RatioSplit

%matplotlib inline

import cornac
from cornac.utils import cache
from cornac.datasets import movielens
from cornac.eval_methods import BaseMethod
from cornac.models import MF, BPR, WMF, IBPR, PMF, WBPR
from cornac.data import Dataset
from cornac.hyperopt import GridSearch, RandomSearch
from cornac.hyperopt import Discrete, Continuous

#%tensorflow_version 1.x
import tensorflow as tf

print(f"System version: {sys.version}")
print(f"Cornac version: {cornac.__version__}")
print(f"Tensorflow version: {tf.__version__}")

SEED = 42
VERBOSE = False
```

```python
# list format
data_list = list(df_rating_withtext_sample.itertuples(index=False, name=None))
# Cornac Dataset Format
dataset = cornac.data.Dataset.from_uir(
    df_rating_withtext_sample.itertuples(index=False), seed=SEED
)
```

```python
ratio_split = RatioSplit(data=data_list, test_size=0.2, verbose=True, seed=SEED)
```

```python
eval_metrics = [
    cornac.metrics.AUC(),
    cornac.metrics.FMeasure(k=50),
    cornac.metrics.Recall(k=50),
    cornac.metrics.NDCG(k=50),
    cornac.metrics.MRR(),
    cornac.metrics.Precision(k=50),
]
```

```python
# 1. base MF model
mf = MF(verbose=VERBOSE, seed=SEED)
cornac.Experiment(
    eval_method=ratio_split, models=[mf], metrics=eval_metrics, user_based=True,
).run()
```

```python
##Tune BPR
# 2. BPR
lrs = [0.1, 0.01, 0.001]
Ks = [30, 50]
lam_regs = [0.01, 0.001]
for K in Ks:
    for lr in lrs:
        for b in lam_regs:
            bpr = cornac.models.BPR(
                k=K,
                max_iter=300,
                learning_rate=lr,
                lambda_reg=b,
                verbose=True,
                seed=SEED,
                name=f"BPR(K={K}) lr={lr} lam={b})",
            )
            cornac.Experiment(
                eval_method=ratio_split,
                models=[bpr],
                metrics=eval_metrics,
                user_based=True,
            ).run()
```

```python
ks = [20, 50]
lrs = [0.1, 0.01, 0.001]
b_s = [0.01, 0.001]
for k in ks:
    for lr in lrs:
         for b in b_s:
            wmf = WMF(
                k=k,
                learning_rate=lr,
                b=b,
                max_iter=300,
                batch_size=8,
                verbose=True,
                seed=SEED,
                name=f"WMF(K={k} lr={lr} b={b})",
            )
            cornac.Experiment(
                eval_method=ratio_split,
                models=[wmf],
                metrics=eval_metrics,
                user_based=True,
            ).run()
```

```python
bpr = cornac.models.BPR(
    k=50, max_iter=300, learning_rate=0.1, lambda_reg=0.01, verbose=True, seed=SEED,
)
cornac.Experiment(
    eval_method=ratio_split, models=[bpr], metrics=eval_metrics, user_based=True,
).run()
```
